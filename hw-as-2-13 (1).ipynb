where

- A is total alcohol consumed, in ounces (oz).
- W is body weight in pounds.
- r is the alcohol distribution ratio:
  - 0.73 for men
  - 0.66 for women
- H is number of hours since the last drink.

Create a program that compares two strings and determines if the two strings are anagrams. The program should prompt for both input strings and display the output as shown in the example that follows.

## Example Output

</pre>

## Constraints

- Implement the program using a function called `isAnagram`, which takes in two words as its arguments and returns `true` or `false`. Invoke this function from your main program.
- Check that both words are the same length.

```python
def isanagram(word1, word2):
    #initialize lists of letters in each word
    word1letters =[]
    word2letters=[]
    
    # return false if lengths are unequal
    if len(word1) != len(word2):
        return False

    else:
        #append letters of each word to the lists
        for letter in word1:
            word1letters.append(letter)
        for letter in word2:
            word2letters.append(letter)
            
        #sort the lists
        word1letters.sort()
        word2letters.sort()
        
        #initalize a counter
        count=0
        
        while count <= len(word2letters)-1:
            # check if indexed letter is the same in both words and then increases the count 
            # until it reaches the end of the words
            if word1letters[count] == word2letters[count]:
                count += 1
                
            #otherwise it's not an anagram
            else:
                return False
            
        #return True if the while loop finishes
        return True
        


            
def problem9():
    print("Enter two strings and I'll tell you if they are anagrams:")
    word1 = input("Enter the first word: ")
    word2 = input("Enter the second word: ")
    
    #check if they're anagrams for final print statement
    if isanagram(word1, word2) == True:
        ag_check = "are"
    else:
        ag_check = "are not"
        
    print('"{}" and "{}" {} anagrams.'.format(word1, word2, ag_check))
        

problem9()
        
```

- A very strong password contains letters, numbers, and special characters and is at least eight characters.

## Example Output

- Create a `passwordValidator` function that takes in the password as its argument and returns a value you can evaluate to determine the password strength. Do not have the function return a string — you may need to support multiple languages in the future.
- Use a single output statement.

```python
def passwordValidator(pw):
    
    # check if length is at least 8
    length = len(str(pw))
    if length < 8:
        return "weak"
    
    # append each character to a list
    pw_chars = []
    for char in pw:
        pw_chars.append(char)
    
    # count all the elements of pw that are integers. if they're all integers, return "weak"
    count_int = 0
    for char in range(0, len(pw_chars)):
        if pw_chars[char].isdigit() == True:
            count_int +=1
    if count_int == len(pw):
        return "weak"
    
    # now do it for characters
    count_str = 0
    for char in range(0, len(pw_chars)):
        if pw_chars[char].isalpha() == True:
            count_str +=1
    if count_str == len(pw):
        return "weak"
    
     # now do it for special characters
    count_special = 0
    for char in range(0, len(pw_chars)):
        if pw_chars[char].isalpha() == False and pw_chars[char].isdigit()== False:
            count_special +=1
    
    if count_int > 0 and count_str > 0 and length >=8:
        if count_special > 0:
            return "very strong"
        return "strong"
    
    
passwordValidator(input())
```

```python
def passwordcreator():
    #prompt for minimum characters, special characters, and number of digits
    length_pw = int(input("What's the length? "))
    special_pw = int(input("How many special characters? "))
    n_pw = int(input("How many numbers? "))
    
    #calc number of letters needed for password
    letters_pw = length_pw - special_pw - n_pw

    #create list of all letters both upper and lowercase
    letters=[]
    for i in range(97,123):
        #convert character code to letter
        l = chr(i)
        
        #append lower and uppercase versions
        low = l.upper()
        up = l.lower()
        letters.append(up)
        letters.append(low)
        
    #create list of all numbers
    numbers=[]
    for i in range(0,10):
        numbers.append(int(i))
    
    #bring in list of special characters
    from string import punctuation
    set(punctuation)
    sc = ['~', ':', '+', '@', '^', '%', '-', '*', '|', 
          '&', '<', '`', '_', '=', '!', '>', ';', '?', '#', '$', '/']
    
    #take random samples of letters, numbers, and specials for password
    import numpy as np
    rand_letters = (np.random.choice(letters, letters_pw, replace=True))
    rand_numbers = (np.random.choice(numbers, n_pw, replace=True))
    rand_special = (np.random.choice(sc, special_pw, replace=True))
    
    #concat all into one array
    pw = np.concatenate((rand_letters, rand_numbers, rand_special))
    
    #shuffle the order
    pw_shuffle = np.random.choice(pw, len(pw), replace=False)
    
    #take from array and create one word 
    final_pw = ''.join(pw_shuffle)
    
    return final_pw
    
passwordcreator()
```

```python
def word_count(file):
    #function to get a list of words sorted by their count in the list

    #read data
    read = open(file, "r")
    data=read.readlines()

    #data reads in as one giant string, so we need to make sure each word is its own string
    words=[]
    for i in data:
        words.append(i)
    data_2 = ",".join(x.strip() for x in words)
    data_3 = data_2.split(" ")

    # use a set to get list of unique names in words
    placeholder_set = set()
    words_unique = []
    for x in data_3:
        placeholder_set.add(x)

    # now convert set into a new list    
    for i in placeholder_set:
        words_unique.append(i)

    #create dictionary, count_dict, to hold counts of each word
    count_dict={}
    for n in range(0, len(words_unique)):
        for i in words_unique:
            count_dict[i] = 0


    #loop through words list for each unique word and count occurrence of each word for the count dict
    for unique in words_unique:
        for i in range(0, len(data_3)):
            if data_3[i] == unique:
                count_dict[unique] += 1

    #Now turn the count_dict into a list with the word counts as asterisks            
    count_stars = []            
    for unique in words_unique:
        count_stars.append(count_dict[unique] * "*")

    #Now turn the words and their star counts into a list of lists to make sorting possible
    final_list=[]
    for i in range(0, len(words_unique)):
        final_list.append([words_unique[i], count_stars[i]])

    # sort the lists based on number of stars
    final_list = sorted(final_list, key=lambda final_list:final_list[1], reverse=True)

    #print the final list of words and their star counts
    for x in final_list:
        print(x[0] + ":", "".join(map(str, x[1:])))


    

my_file="C:\\Users\\ja041718\\Downloads\\problem30data.txt"        
word_count(my_file)
```
